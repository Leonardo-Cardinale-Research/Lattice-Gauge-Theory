```python
import numpy as np
from scipy.linalg import sqrtm

def generate_su2_matrices(N):
    matrices = np.zeros((N, 2, 2), dtype=np.complex128)

    # Generate random 2x2 complex matrices
    random_mat = np.random.rand(N, 2, 2) + 1j * np.random.rand(N, 2, 2)

    # Convert the matrices to Hermitian matrices
    hermitian_mat = 0.5 * (random_mat + np.conj(random_mat))
    
    print(hermitian_mat.shape)

    # Compute the square root of the Hermitian matrices
    sqrt_hermitian_mat = sqrtm(hermitian_mat)

    # Compute the determinant of the square root matrices
    det = np.linalg.det(sqrt_hermitian_mat)

    # Normalize the matrices to have unit determinant
    matrices = sqrt_hermitian_mat / np.power(det[..., np.newaxis, np.newaxis], 0.25)

    return matrices

# Set the number of matrices to generate
N = 5

# Generate N random SU(2) matrices
su2_matrices = generate_su2_matrices(N)
```
